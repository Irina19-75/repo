```python
#q.Получить сумму всех значений в стартовой матрице mat

import numpy as np

mat = np.arange(1,26).reshape(5,5)
print(mat)
print('-------------------')
print(mat.sum())
```

```python
#r.Получить стандартное отклонение в стартовой матрице mat

import numpy as np

mat = np.arange(1,26).reshape(5,5)
print(mat)
print('-------------------')
result = np.std(mat)
print('SD:', result)
```

```python
#s.Получить сумму значений в колонках в стартовой матрице mat


import numpy as np

mat = np.arange(1,26).reshape(5,5)
print(mat)
print('-------------------')
# To sum over columns:
print('To sum over columns:',(mat.sum(axis=0)))
print('--------------------')
# Or, to sum over rows:
print('Or, to sum over rows:',(mat.sum(axis=1)))

```
